```python
#libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Global Thresholding

```python
def threshold_segmentation(image_path, threshold_value):
    # Load the image
    image = cv2.imread(image_path, 0)  # Load image in grayscale

    # Apply thresholding
    _, thresholded = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

    # Display the original and thresholded images using Matplotlib
    fig, axes = plt.subplots(1, 2,figsize=(15, 8))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title('Global Thresholded ')
    axes[1].axis('off')
    plt.show()
    # using
image_path = "cat.jpg"
threshold_value = 128
threshold_segmentation(image_path, threshold_value)
```

## Adaptive Thresholding

```python
def apply_adaptive_threshold(image, block_size, constant):
    # Apply adaptive thresholding
    thresholded = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
    return thresholded

def display_thresholded_image(image, block_size, constant):
    # Apply adaptive thresholding
    thresholded = apply_adaptive_threshold(image, block_size, constant)

    # Display the original and thresholded images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title('Adaptive Thresholding')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# using
image_path = "cat.jpg"
image = cv2.imread(image_path, 0)  # Load image in grayscale

block_size = 13 # Neighborhood block size for adaptive thresholding
constant = 2     # Constant subtracted from the mean

display_thresholded_image(image, block_size, constant)
```

## Otsu's Thresholding

```python
def apply_otsu_threshold(image):
    # Apply Otsu's thresholding
    _, thresholded = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return thresholded

def display_thresholded_image(image):
    # Apply Otsu's thresholding
    thresholded = apply_otsu_threshold(image)

    # Display the original and thresholded images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title("Otsu's Thresholding")
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# using
image_path = "cat.jpg"
image = cv2.imread(image_path, 0)  # Load image in grayscale

display_thresholded_image(image)
```

## Multi-Level Thresholding

```python

def apply_multi_level_threshold(image, threshold_values):
    # Convert image to grayscale if needed
    if len(image.shape) > 2:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply multi-level thresholding
    _, thresholded = cv2.threshold(image, threshold_values[0], 255, cv2.THRESH_BINARY)
    for i in range(len(threshold_values) - 1):
        _, temp = cv2.threshold(image, threshold_values[i+1], 255, cv2.THRESH_BINARY)
        thresholded = cv2.bitwise_or(thresholded, temp)

    return thresholded

def display_thresholded_image(image, threshold_values):
    # Apply multi-level thresholding
    thresholded = apply_multi_level_threshold(image, threshold_values)

    # Display the original and thresholded images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title('Multi-Level Thresholding')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# using
image_path = "cat.jpg"
image = cv2.imread(image_path)  # Load image

threshold_values = [100, 150, 200]  # Adjust threshold values as needed

display_thresholded_image(image, threshold_values)
```

## Inverse Thresholding

```python


def apply_inverse_threshold(image, threshold_value):
    # Apply inverse thresholding
    _, thresholded = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY_INV)
    return thresholded

def display_thresholded_image(image, threshold_value):
    # Apply inverse thresholding
    thresholded = apply_inverse_threshold(image, threshold_value)

    # Display the original and thresholded images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title('Inverse Thresholding')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# using
image_path = "cat.jpg"
image = cv2.imread(image_path, 0)  # Load image in grayscale

threshold_value = 128  # Adjust threshold value as needed

display_thresholded_image(image, threshold_value)
```

## Truncated Thresholding

```python


def apply_truncated_threshold(image, threshold_value):
    # Apply truncated thresholding
    _, thresholded = cv2.threshold(image, threshold_value, threshold_value, cv2.THRESH_TRUNC)
    return thresholded

def display_thresholded_image(image, threshold_value):
    # Apply truncated thresholding
    thresholded = apply_truncated_threshold(image, threshold_value)

    # Display the original and thresholded images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(thresholded, cmap='gray')
    axes[1].set_title('Truncated Thresholding')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# using
image_path = "cat.jpg"
image = cv2.imread(image_path, 0)  # Load image in grayscale

threshold_value = 128  # Adjust threshold value as needed

display_thresholded_image(image, threshold_value)
```
